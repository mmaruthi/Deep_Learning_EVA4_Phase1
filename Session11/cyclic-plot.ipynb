```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

a=0
b=5

x = [a,a+b,a+2*b,a+3*b,a+4*b]
y = [a,a+b,a,a+b,a]


plt.plot(x, y, label = "Cyclic Curve")
x = np.linspace(0, 10, 1000)
ax = plt.axes()

# Set the y axis label of the current axis.
plt.xlabel('x - axis')

# Set the y axis label of the current axis.
plt.ylabel('y - axis')

# show a legend on the plot
plt.legend()
# Display a figure.
plt.show()
```
